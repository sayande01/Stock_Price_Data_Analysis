# S&P 500 Companies Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set()
```

```python
# Importing dataset
```

```python
df = pd.read_csv('financials.csv')
```

### Expolaratory Data Analysis

```python
#top rows
df.head()
```

```python
#Row-Column
df.shape
```

```python
#Details
df.info()
```

```python
#Columns
df.columns
```

```python
#Statistics
df.describe()
```

```python
#Null check
df.isnull().sum()
```

```python
# Filling the null values with mean values of the column
df['Price/Earnings'].fillna(df['Price/Earnings'].mean(), inplace = True)
df['Price/Book'].fillna(df['Price/Book'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
# Top companies with most market share
```

```python
df1 = df[['Name', 'Market Cap']].sort_values( by = 'Market Cap', ascending = False).head(10)
df1
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Name', y = 'Market Cap', data = df1)
plt.xticks(rotation = 45)
plt.show()
```

```python
# Top Share Prices
```

```python
df2 = df[['Name', 'Price']].sort_values( by = 'Price', ascending = False).head(10)
df2
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'Price', y = 'Name', data = df2)
plt.ylabel(None)
plt.show()
```

```python
#Sector wise
```

```python
df3 = df['Sector'].value_counts().head(5)
df3
```

```python
df3.plot(kind = 'pie', autopct = '%1.2f%%')
plt.ylabel(None)
plt.show()
```

```python
# Sector wise avg shares price
```

```python
df4 = df.groupby('Sector')['Price'].mean().sort_values(ascending = True)
df4
```

```python
df4.plot()
plt.xticks(rotation = 45)
plt.show()
```

```python
# Violin Plot
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x = 'Sector', y = 'Price', data = df)
plt.xticks(rotation = 45)
plt.show()
```

```python
# Share price dist
```

```python
df['Price'].hist()
plt.show()
```

# USA Top 50 Companies Data Analysis

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Top 10 Comp with largest market Cap
```

```python
df7 = data[['Company Name','Market Cap (USD in Trillions)']].sort_values(by = 'Market Cap (USD in Trillions)', ascending = False).head(10)
df7
```

```python
sns.barplot(x = 'Market Cap (USD in Trillions)', y = 'Company Name', data = df7)
plt.show()
```

All the companies in the top50 are of 'Technology' sector

```python
df8 = data['Sector'].value_counts()
df8
```

```python
# Sector wise No of Companies
```

```python
df8.plot(kind = 'pie')
plt.ylabel(None)
plt.show()
```

```python
df9 = data['HQ State'].value_counts().head(5)
df9
```

```python
# State head Quarter
```

```python
df9.plot(kind= 'bar')
plt.xticks(rotation = 60)
plt.xlabel(None)
plt.show()
```

```python
# Top companies wrt Employee count
```

```python
df10 = data[['Company Name','Employee Size']].sort_values(by = 'Employee Size', ascending = False).head(10)
df10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'Company Name', y = 'Employee Size', data = df10)
plt.xticks(rotation = 60)
plt.show()
```

```python
# Founding Year wise Company Count
```

```python
sns.histplot(x = 'Founding Year', data = data, bins = 20, kde = True)
plt.show()
```

```python
# Sector wise Annual revenue distribution
```

```python
df12 = data.groupby('Sector')['Annual Revenue 2022-2023 (USD in Billions)'].mean().sort_values(ascending = False)
df12
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = 'Sector', y = 'Annual Revenue 2022-2023 (USD in Billions)', data = data)
plt.xticks(rotation = 45)
plt.show()
```

```python
# Highest tax paying comp's
```

```python
df13 = data[['Company Name','Annual Income Tax in 2022-2023 (USD in Billions)']].sort_values(by = 'Annual Income Tax in 2022-2023 (USD in Billions)', ascending = False).head()
df13
```

```python
sns.barplot(x = 'Annual Income Tax in 2022-2023 (USD in Billions)', y = 'Company Name', data = df13)
plt.title('Annual Income Tax in 2022-2023 (USD in Billions)')
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```
